```python
import pandas as pd
import numpy as np
import altair as alt
import re
from altair_saver import save
```

```python
public = pd.read_csv('data/NCES_public_clean.csv', index_col=0)
public = public.loc[:, ['School Name [Public School] 2015-16', 'State Name [Public School] 2015-16', 'Location City [Public School] 2015-16', 'Urban-centric Locale [Public School] 2015-16', 'Pupil/Teacher Ratio [Public School] 2015-16', 'American Indian/Alaska Native Students [Public School] 2015-16', 'Asian or Asian/Pacific Islander Students [Public School] 2015-16', 'Hispanic Students [Public School] 2015-16', 'Black or African American Students [Public School] 2015-16', 'White Students [Public School] 2015-16', 'Total Race/Ethnicity [Public School] 2015-16']]
public.columns = ['School Name', 'State Name', 'City Name', 'Urban-centric Locale', 'Pupil/Teacher Ratio', 'Native American', 'Asian', 'Hispanic', 'Black', 'White', 'Total']

public = public.dropna()
public = public.astype({'Native American': str, 'Asian': str, 'Hispanic': str, 'Black': str, 'White': str, 'Total': str, 'Pupil/Teacher Ratio': str, 'Urban-centric Locale': str})
native_american = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['Native American']])
asian = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['Asian']])
hispanic = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['Hispanic']])
black = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['Black']])
white = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['White']])
ratio = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['Pupil/Teacher Ratio']])
urban = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in public['Urban-centric Locale']])
public = public[native_american & asian & hispanic & black & white & ratio & urban]

public = public.astype({'Native American': float, 'Asian': float, 'Hispanic': float, 'Black': float, 'White': float, 'Total': float, 'Pupil/Teacher Ratio': float})
public['Percentage of Native American'] = public['Native American'] / public['Total'] * 100
public['Percentage of Asian'] = public['Asian'] / public['Total'] * 100
public['Percentage of Hispanic'] = public['Hispanic'] / public['Total'] * 100
public['Percentage of Black'] = public['Black'] / public['Total'] * 100
public['Percentage of White'] = public['White'] / public['Total'] * 100

public = public.loc[:, ['School Name', 'State Name', 'City Name', 'Urban-centric Locale', 'Pupil/Teacher Ratio', 'Native American', 'Asian', 'Hispanic', 'Black', 'White', 'Percentage of Native American', 'Percentage of Asian', 'Percentage of Hispanic', 'Percentage of Black', 'Percentage of White']]
public = public[public['State Name'] == 'NEW YORK']

public['School'] = 'public'

public.head()
```

```python
private = pd.read_csv('data/NCES_private_clean.csv', index_col=0)
private = private.loc[:, ['Private School Name [Private School] 2015-16', 'State Name [Private School] 2015-16', 'City [Private School] 2015-16', 'Urban-centric Locale [Private School] 2015-16', 'Pupil/Teacher Ratio [Private School] 2015-16', 'American Indian/Alaska Native Students [Private School] 2015-16', 'Asian or Asian/Pacific Islander Students [Private School] 2015-16', 'Hispanic Students [Private School] 2015-16', 'Black or African American Students [Private School] 2015-16', 'White Students [Private School] 2015-16', 'Percentage of American Indian/Alaska Native Students [Private School] 2015-16', 'Percentage of Asian or Asian/Pacific Islander Students [Private School] 2015-16', 'Percentage of Hispanic Students [Private School] 2015-16', 'Percentage of Black Students [Private School] 2015-16', 'Percentage of White Students [Private School] 2015-16']]
private.columns = ['School Name', 'State Name', 'City Name', 'Urban-centric Locale', 'Pupil/Teacher Ratio', 'Native American', 'Asian', 'Hispanic', 'Black', 'White', 'Percentage of Native American', 'Percentage of Asian', 'Percentage of Hispanic', 'Percentage of Black', 'Percentage of White']

private = private.dropna()
private = private.astype({'Native American': str, 'Asian': str, 'Hispanic': str, 'Black': str, 'White': str, 'Pupil/Teacher Ratio': str, 'Urban-centric Locale': str})
native_american = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['Native American']])
asian = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['Asian']])
hispanic = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['Hispanic']])
black = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['Black']])
white = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['White']])
ratio = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['Pupil/Teacher Ratio']])
urban = np.array([False if re.match('[^a-zA-Z0-9.\-: ]+', cell) else True for cell in private['Urban-centric Locale']])
private = private[native_american & asian & hispanic & black & white & ratio & urban]

private = private.astype({'Native American': float, 'Asian': float, 'Hispanic': float, 'Black': float, 'White': float, 'Percentage of Native American': float, 'Percentage of Asian': float, 'Percentage of Hispanic': float, 'Percentage of Black': float, 'Percentage of White': float, 'Pupil/Teacher Ratio': float})

private = private[private['State Name'] =='NEW YORK']

private['School'] = 'private'

private.head()
```

```python
data = pd.concat([public,private])
data.head()
```

```python
alt.data_transformers.enable('json')
```

```python
points_public = alt.Chart(data).transform_filter(
    (alt.datum.School == 'public')
).mark_point(size=50, filled=True, opacity=0.5).encode(
    x=alt.X('Percentage of Black'),
    y=alt.Y('Percentage of White'),
    color = 'School'
)

points_private = alt.Chart(data).transform_filter(
    (alt.datum.School == 'private')
).mark_point(size=50, filled=True, opacity=0.5).encode(
    x=alt.X('Percentage of Black'),
    y=alt.Y('Percentage of White'),
    color = 'School'
)

points_public_median = alt.Chart(data).transform_filter(
    (alt.datum.School == 'public')
).mark_point(size=80, filled=True, color='red').encode(
    x=alt.X('median(Percentage of Black)'),
    y=alt.Y('median(Percentage of White)'),
)

points_private_median = alt.Chart(data).transform_filter(
    (alt.datum.School == 'private')
).mark_point(size=80, filled=True, color='aquamarine').encode(
    x=alt.X('median(Percentage of Black)'),
    y=alt.Y('median(Percentage of White)'),
)

distribright_public = alt.Chart(data).transform_filter(
    (alt.datum.School == 'public')
).mark_bar().encode(
    y = alt.Y('Percentage of White',bin=alt.BinParams(maxbins=20),axis=None),
    x = alt.X('count()',axis=None),
    color = 'School'
).properties(width=30)

distribright_private = alt.Chart(data).transform_filter(
    (alt.datum.School == 'private')
).mark_bar().encode(
    y = alt.Y('Percentage of White',bin=alt.BinParams(maxbins=20),axis=None),
    x = alt.X('count()',axis=None),
    color = 'School'
).properties(width=30)

distribtop_public = alt.Chart(data).transform_filter(
    (alt.datum.School == 'public')
).mark_bar().encode(
    x = alt.X('Percentage of Black',bin=alt.BinParams(maxbins=20),axis=None),
    y = alt.Y('count()',axis=None),
    color = 'School'
).properties(height=30)

distribtop_private = alt.Chart(data).transform_filter(
    (alt.datum.School == 'private')
).mark_bar().encode(
    x = alt.X('Percentage of Black',bin=alt.BinParams(maxbins=20),axis=None),
    y = alt.Y('count()',axis=None),
    color = 'School'
).properties(height=30)

chart1 = ((distribtop_public & ((points_public + points_public_median) | distribright_public)).properties(title='Share of White and Black Student in Public Schools in New York State')) | ((distribtop_private & ((points_private + points_private_median) | distribright_private)).properties(title='Share of White and Black Student in Private Schools in New York State'))

chart1
```

```python
def label_location(x):
    if 'City' in x:
        return 'City'
    elif 'Suburb' in x:
        return 'Suburb'
    elif 'Town' in x:
        return 'Town'
    else:
        return 'Rural'
```

```python
data['Location'] = data['Urban-centric Locale'].apply(label_location)
data['Teacher/Pupil Ratio'] = 1 / data['Pupil/Teacher Ratio']
data['Dominant Race'] = data.loc[:, ['Native American', 'Asian', 'Hispanic', 'Black', 'White']].idxmax(axis=1)
data.head()
```

```python
bar = alt.Chart(data).mark_bar(color='orange').encode(
    x=alt.X('mean(Teacher/Pupil Ratio):Q'),
    y=alt.Y('Dominant Race:N')
).properties(height=300)

line = alt.Chart(data).mark_line(color='red').encode(
    x=alt.X('median(Teacher/Pupil Ratio):Q'),
    y=alt.Y('Dominant Race:N')
).properties(height=300)

text1 = bar.mark_text(
    align='left',
    baseline='middle',
    dx=5
).encode(
    text=alt.Text('mean(Teacher/Pupil Ratio):Q', format='s')
)

text2 = line.mark_text(
    align='right',
    baseline='middle',
    dx=-15,
    color='white'
).encode(
    text=alt.Text('median(Teacher/Pupil Ratio):Q', format='s')
)

heatmap = alt.Chart(data).mark_rect().encode(
    x=alt.X('Location:N', sort=alt.SortArray(
        ['Rural', 'Town', 'Suburb', 'City']
        ), axis=alt.Axis(labelAngle=0)),
    y=alt.Y('Dominant Race:N', axis=None),
    color='count():Q'
).properties(height=300, width=300)

text3 = alt.Chart(data).mark_text(
    align='center',
    baseline='middle'
).transform_joinaggregate(
    count='count()',
    groupby=['Location', 'Dominant Race']
).encode(
    x=alt.X('Location:N', sort=alt.SortArray(
        ['Rural', 'Town', 'Suburb', 'City']
        )),
    y=alt.Y('Dominant Race:N', axis=None),
    text='count:Q',
    color=alt.condition(
        alt.datum.count > 950,
        alt.value('white'),
        alt.value('black')
    )
).properties(height=300, width=300)

highlight = alt.Chart(data).transform_joinaggregate(
    count='count()',
    groupby=['Location', 'Dominant Race']
).transform_filter(
    alt.datum.count > 950
).mark_rect(
    color='', 
    stroke='red', 
    strokeWidth=2
).encode(
    x=alt.X('Location:N', sort=alt.SortArray(
        ['Rural', 'Town', 'Suburb', 'City']
        )),
    y=alt.Y('Dominant Race:N', axis=None)
).properties(height=300, width=300)

chart2 = ((bar + line + text1 + text2) | (heatmap + text3 + highlight)).resolve_scale(y='shared').properties(title='Educational Resources and Locations of Schools by Dominant Race in New York State')

chart2
```

```python
math = pd.read_csv('data/ny-math-results-2013-2019-public-all.csv', index_col=0)
math = math[math['Grade'] == 'All Grades']
math = math.loc[:, ['School Name', 'Year', 'Mean Scale Score']]
math = math.dropna()

ela = pd.read_csv('data/ny-ela-results-2013-2019-public-all.csv', index_col=0)
ela = ela[ela['Grade'] == 'All Grades']
ela = ela.loc[:, ['School Name', 'Year', 'Mean Scale Score']]
ela = ela.dropna()
```

```python
black_school = data.sort_values(by=['Black'], ascending=False)['School Name']
black_math = pd.DataFrame([], columns=['School Name', 'Year', 'Mean Scale Score', 'Rank'])
black_ela = pd.DataFrame([], columns=['School Name', 'Year', 'Mean Scale Score', 'Rank'])

rank = 1
for school in black_school:
    if school in math['School Name'].values:
        add_math = math[math['School Name'] == school]
        add_math['Rank'] = rank
        black_math = black_math.append(add_math, ignore_index=True)
        add_ela = ela[ela['School Name'] == school]
        add_ela['Rank'] = rank
        black_ela = black_ela.append(add_ela, ignore_index=True)
        rank += 1
    if rank > 5:
        break
```

```python
white_school = data.sort_values(by=['White'], ascending=False)['School Name']
white_math = pd.DataFrame([], columns=['School Name', 'Year', 'Mean Scale Score', 'Rank'])
white_ela = pd.DataFrame([], columns=['School Name', 'Year', 'Mean Scale Score', 'Rank'])

rank = 1
for school in white_school:
    if school in math['School Name'].values:
        add_math = math[math['School Name'] == school]
        add_math['Rank'] = rank
        white_math = white_math.append(add_math, ignore_index=True)
        add_ela = ela[ela['School Name'] == school]
        add_ela['Rank'] = rank
        white_ela = white_ela.append(add_ela, ignore_index=True)
        rank += 1
    if rank > 5:
        break
```

```python
black_math['Race'] = 'black'
black_math['Subject'] = 'math'
black_ela['Race'] = 'black'
black_ela['Subject'] = 'language'
white_math['Race'] = 'white'
white_math['Subject'] = 'math'
white_ela['Race'] = 'white'
white_ela['Subject'] = 'language'

score = pd.concat([black_math,black_ela,white_math,white_ela])
score.head()
```

```python
base1 = alt.Chart(score).transform_filter(
    (alt.datum.Year == 2013) & (alt.datum.Subject == 'math')
).encode(
    x=alt.X('Mean Scale Score:Q', axis=alt.Axis(title=None)),
    y=alt.Y('Race:O', axis=alt.Axis(labels=False, title=None)),
    color=alt.Color('Race:O', scale=alt.Scale(scheme='category10'))
)

layer1 = alt.layer(
  base1.mark_bar(),
  base1.mark_text(dx=25).encode(text=alt.Text('Mean Scale Score:Q', format='s'))
).facet(
  row='Rank:O'
).properties(title='Year 2013')

base2 = alt.Chart(score).transform_filter(
    (alt.datum.Year == 2019) & (alt.datum.Subject == 'math')
).encode(
    x=alt.X('Mean Scale Score:Q', axis=alt.Axis(title=None)),
    y=alt.Y('Race:O', axis=alt.Axis(labels=False, title=None)),
    color='Race:O'
)

layer2 = alt.layer(
  base2.mark_bar(),
  base2.mark_text(dx=25).encode(text=alt.Text('Mean Scale Score:Q', format='s'))
).facet(
  row=alt.Row('Rank:O', header=alt.Header(labels=False), title=None)
).properties(title='Year 2019')

base3 = alt.Chart(score).transform_filter(
    (alt.datum.Year == 2013) & (alt.datum.Subject == 'language')
).encode(
    x=alt.X('Mean Scale Score:Q', axis=alt.Axis(title=None)),
    y=alt.Y('Race:O', axis=alt.Axis(labels=False, title=None)),
    color='Race:O'
)

layer3 = alt.layer(
  base3.mark_bar(),
  base3.mark_text(dx=25).encode(text=alt.Text('Mean Scale Score:Q', format='s'))
).facet(
  row='Rank:O'
).properties(title='Year 2013')

base4 = alt.Chart(score).transform_filter(
    (alt.datum.Year == 2019) & (alt.datum.Subject == 'language')
).encode(
    x=alt.X('Mean Scale Score:Q', axis=alt.Axis(title=None)),
    y=alt.Y('Race:O', axis=alt.Axis(labels=False, title=None)),
    color='Race:O'
)

layer4 = alt.layer(
  base4.mark_bar(),
  base4.mark_text(dx=25).encode(text=alt.Text('Mean Scale Score:Q', format='s'))
).facet(
  row=alt.Row('Rank:O', header=alt.Header(labels=False), title=None)
).properties(title='Year 2019')

chart3 = (((layer1 | layer2).properties(title='Mean of Math')) & ((layer3 | layer4).properties(title='Mean of Language'))).properties(title='Scores in New York Public Schools with Highest Number of Black or Whites')

chart3
```
